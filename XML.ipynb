### From the contents of both variables, form lists C1 and C2, in which the contents of the XML data are stored as dictionary-type objects. **NOTE**: the attributes presented at the record level must be included in the objects and the identifiers under the address identifier must be brought to a higher level next to user's other properties!

Below is a model showing the structure of the first object:

<pre>[{
    'type': 'customer',
    'name': 'Morgan Sykes',
    'phone': '1-183-546-6564',
    'email': 'lorem.auctor@vulputatemauris.net',
    'streetaddress' : '2594 Tellus St.',
    'country': 'Russian Federation',
    'postalZip': '584585'
}]<pre>



```python
records_from_string = root_from_string.findall('record')
records_from_file = root_from_file.findall('record')

C1 = []
C2 = []

for record in records_from_string + records_from_file:
    data = {
        'name': record.find('name').text,
        'phone': record.find('phone').text if record.find('phone') is not None else None,
        'email': record.find('email').text if record.find('email') is not None else None,
        'type': record.attrib['type'],
        'streetaddress': record.find('address/streetaddress').text if record.find('address/streetaddress') is not None else None,
        'country': record.find('address/country').text if record.find('address/country') is not None else None,
        'postalZip': record.find('address/postalZip').text if record.find('address/postalZip') is not None else None
    }

    if data['type'] == 'customer':
        C1.append(data)
    else:
        C2.append(data)

C1[:3]
C2[:3]
```

### Convert both lists into dataframes and finally combine them into one dataframe (remember to reset the index of the new dataframe!). Print the content of the generated dataframe.

```python
df_C1 = pd.DataFrame(C1)
df_C2 = pd.DataFrame(C2)

combined_df = pd.concat([df_C1, df_C2], ignore_index=True)

combined_df
```

### Some users are missing a phone number. Print the number of these users.

```python
missing_phone_count = combined_df['phone'].isna().sum()
missing_phone_count
```
